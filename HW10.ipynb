```python
# По данным файла power.csv посчитайте суммарное потребление стран Прибалтики (Латвия, Литва и Эстония) 
# категорий 4, 12 и 21 за период с 2005 по 2010 года. Не учитывайте в расчетах отрицательные значения quantity.
```

```python
power = pd.read_csv('power.csv')
```

```python
power.head()
```

```python
country_stats_positive = power.query('quantity > ={}'.format('0'))
country_stats_positive.head()
```

```python
filtered_country_stats = country_stats_positive[ (country_stats_positive['category']==int('4'))| 
                                                (country_stats_positive['category']==int('12')) |
                                                (country_stats_positive['category']==int('21')) ]
```

```python
filtered_by_year = filtered_country_stats[ (filtered_country_stats['year']==int('2005'))| 
                                            (filtered_country_stats['year']==int('2006'))| 
                                            (filtered_country_stats['year']==int('2007'))| 
                                            (filtered_country_stats['year']==int('2008'))| 
                                            (filtered_country_stats['year']==int('2009'))| 
                                            (filtered_country_stats['year']==int('2010'))  ]
```

```python
filtered_by_year.head()
```

```python
filtered_by_country = filtered_by_year[ (filtered_by_year['country']==('Lithuania'))| 
                                            (filtered_by_year['country']==('Latvia'))| 
                                            (filtered_by_year['country']==('Estonia')) 
                                              ]
```

```python
filtered_by_country.head()
```

```python
result = sum(filtered_by_country['quantity'])
```

```python
print(result)
```
